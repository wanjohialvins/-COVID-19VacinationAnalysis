```python
# Import the pandas library and give it an alias 'pd'
import pandas as pd

# Import the numpy library and give it an alias 'np'
import numpy as np

# Import the seaborn library and give it an alias 'sns'
import seaborn as sns

# Import the matplotlib.pyplot library and give it an alias 'plt'
import matplotlib.pyplot as plt
```

```python
# Read data from a CSV file using pandas. The file is named 'Sample - Superstore.csv'.
# The 'encoding' parameter is set to 'latin-1' to handle non-UTF-8 encoded characters.
df = pd.read_csv('Sample - Superstore.csv', encoding='latin-1')

# Display the first few rows (default is 5) of the DataFrame 'df' to inspect the data.
df.head()
```

```python
# Obtain information about the DataFrame 'df' and its data distribution.
df.info()
```

```python
# Generate summary statistics for the data in the DataFrame 'df'.
df.describe()
```

```python
#Visualization of Shipmode, Segment and City
# Create a figure with a size of 12x8 inches.
plt.figure(figsize=(12, 8))

# Create a subplot grid of 2 rows and 2 columns, and select the first subplot.
plt.subplot(2, 2, 1)

# Create a countplot to visualize the distribution of 'Ship Mode'.
sns.countplot(data=df, x='Ship Mode')
plt.title('Ship Mode Distribution')

# Select the second subplot.
plt.subplot(2, 2, 2)

# Create a countplot to visualize the distribution of 'Segment'.
sns.countplot(data=df, x='Segment')
plt.title('Segment Distribution')

# Select the third subplot.
plt.subplot(2, 2, 3)

# Create a countplot to visualize the top 10 cities by orders.
# The 'order' parameter ensures that the cities are ordered by the number of orders.
sns.countplot(data=df, y='City', order=df['City'].value_counts().index[:10])
plt.title('Top 10 Cities by Orders')

# Automatically adjust subplot positions for better layout.
plt.tight_layout()

# Display the visualizations.
plt.show()
```

```python
# Visualization of Sales,Quanityt,Discount, and Profit 
# Create a new figure with a size of 12x8 inches.
plt.figure(figsize=(12, 8))

# Select the first subplot.
plt.subplot(2, 2, 1)

# Create a histogram (histplot) to visualize the distribution of 'Sales' with 20 bins.
sns.histplot(data=df, x='Sales', bins=20)
plt.title('Sales Distribution')

# Select the second subplot.
plt.subplot(2, 2, 2)

# Create a histogram (histplot) to visualize the distribution of 'Quantity' with 20 bins.
sns.histplot(data=df, x='Quantity', bins=20)
plt.title('Quantity Distribution')

# Select the third subplot.
plt.subplot(2, 2, 3)

# Create a histogram (histplot) to visualize the distribution of 'Discount' with 20 bins.
sns.histplot(data=df, x='Discount', bins=20)
plt.title('Discount Distribution')

# Select the fourth subplot.
plt.subplot(2, 2, 4)

# Create a histogram (histplot) to visualize the distribution of 'Profit' with 20 bins.
sns.histplot(data=df, x='Profit', bins=20)
plt.title('Profit Distribution')

# Automatically adjust subplot positions for better layout.
plt.tight_layout()

# Display the visualizations.
plt.show()
```

```python
#Data splitting and subset creation
# Create subsets for cities 'Philadelphia' and 'Los Angeles' with quantities greater than 5.
subset1 = df[(df['City'] == 'Philadelphia') & (df['Quantity'] > 5)]
subset2 = df[(df['City'] == 'Los Angeles') & (df['Quantity'] > 5)]
```

```python
#Summary Statistics calculations
# Calculate summary statistics for the sales distribution in each subset.
summary_subset1 = subset1['Sales'].describe()
summary_subset2 = subset2['Sales'].describe()
```

```python
#Visualization of Sales Distribution
# Create a figure with a size of 12x6 inches.
plt.figure(figsize=(12, 6))

# Create subplots for visualizing the sales distribution in each subset.
plt.subplot(1, 2, 1)
sns.histplot(data=subset1, x='Sales', bins=20)
plt.title('Sales Distribution in Philadelphia (Quantity > 5)')

plt.subplot(1, 2, 2)
sns.histplot(data=subset2, x='Sales', bins=20)
plt.title('Sales Distribution in Los Angeles (Quantity > 5)')

# Automatically adjust subplot positions for better layout.
plt.tight_layout()

# Display the visualizations.
plt.show()
```

```python
#Display Summary Statistics
# Display summary statistics for each subset.
print("Summary Statistics for Philadelphia Subset:")
print(summary_subset1)
print("Summary Statistics for Los Angeles Subset:")
print(summary_subset2)
```

```python
#Data preparation and correlations
# Step 1: Filter the DataFrame to include only numeric columns
numeric_columns = df.select_dtypes(include='number')

# Calculate the correlation matrix
correlation_matrix = numeric_columns.corr()
```

```python
#Identify Features with the strongest correlations
# Step 2: Identify the features with the strongest positive and negative correlations with 'Sales'
# Filter out the 'Sales' column
correlation_with_sales = correlation_matrix['Sales'].drop('Sales')

# Find the feature with the strongest positive correlation
strongest_positive_corr_feature = correlation_with_sales.idxmax()

# Find the feature with the strongest negative correlation
strongest_negative_corr = correlation_matrix['Sales'].nsmallest(1).index
```

```python
#Create Scatterplots to visualize relationships
# Step 3: Create plots to represent these relationships
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x=strongest_positive_corr_feature, y='Sales')
plt.title(f'Positive Correlation: {strongest_positive_corr_feature} vs. Sales')

plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x=strongest_negative_corr[0], y='Sales')
plt.title(f'Negative Correlation: {strongest_negative_corr[0]} vs. Sales')

# Automatically adjust subplot positions for better layout.
plt.tight_layout()

# Display the scatterplots.
plt.show()
```

```python
# Step 1: Choose two significant columns
col1 = 'Profit'
col2 = 'Discount'

# Step 2: Create the new feature by subtracting one column from the other
df['Net_profit'] = df[col1] - df[col2]

# Step 3: Visualize the relationship between the new feature and sales
plt.scatter(df['Net_profit'], df['Sales'])
plt.xlabel('New Profit')
plt.ylabel('Sales')
plt.title('Relationship between Net profit and Sales')
plt.show()

```
